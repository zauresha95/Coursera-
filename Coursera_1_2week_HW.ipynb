# Аппроксимация функции


```python
import matplotlib.pyplot as plt
%matplotlib inline
import math
from scipy import linalg
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)


a=np.array([[1,1],[1,15]])
b=np.array([f(1),f(15)])

w=linalg.solve(a,b)

def f_app(x):
    return w[0]+w[1]*x

from matplotlib import pylab as plt

x = np.arange(1, 15, 0.001)
y = f(x)
y2=f_app(x)
plt.plot(x, y,'r',x,y2,'g')
plt.show()

w
```
